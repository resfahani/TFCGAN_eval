```python
mag = [ 4, 4.5, 5, 5.5, 6]
dis = [30, 50, 70, 90, 110]

fig, axes = plt.subplots(5, 5, figsize =(18,10))

for i in range(len(mag)):
    axy = axes[i]
    for ii in range(len(dis)):
        axx = axy[ii]
        axx.set_yticks([])
        axx.set_xticks([])

        axx.set(frame_on=False)
        
        freq , xh, S, x = Model.Maker(mag[i], dis[ii], 400, 1)
        axx.imshow(np.log10(S[0][-20::-1,:]),aspect='auto', cmap='RdYlGn_r',clim=[-10,3],  extent = [0,1, 0, 1],)
        
        axx.plot(np.linspace(0,1,len(x)), ((x)/ (np.max(np.abs(x))*10))-0.1 , 'k')

        

axes[0][0].set_title("R$_{hyp}$=" + str(dis[0])+"km", fontsize=16)
axes[0][1].set_title("R$_{hyp}$=" + str(dis[1])+"km", fontsize=16)
axes[0][2].set_title("R$_{hyp}$=" + str(dis[2])+"km", fontsize=16)
axes[0][3].set_title("R$_{hyp}$=" + str(dis[3])+"km", fontsize=16)
axes[0][4].set_title("R$_{hyp}$=" + str(dis[4])+"km", fontsize=16)

axes[0][0].set_ylabel("M$_{w}$ = "+ str(mag[0]), fontsize=16)
axes[1][0].set_ylabel("M$_{w}$ = "+ str(mag[1]), fontsize=16)
axes[2][0].set_ylabel("M$_{w}$ = "+ str(mag[2]), fontsize=16)
axes[3][0].set_ylabel("M$_{w}$ = "+ str(mag[3]), fontsize=16)
axes[4][0].set_ylabel("M$_{w}$ = "+ str(mag[4]), fontsize=16)

plt.savefig("FigureTFR.png", dpi = 400,bbox_inches='tight')
```
